```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%pylab inline

font = {'size'   : 18}
plt.rc('font', **font)
```

```python
CPI_data = pd.read_csv('CPIAUCSL.csv', sep=',')
RS_data = pd.read_csv('RSAFS.csv', sep=',')
```

```python
CPI_data['DATE'] = pd.to_datetime(CPI_data['DATE'], format = '%Y.%m.%d')     #  
CPI_data.rename(columns= {'DATE': u'', 'CPIAUCSL_PCH': 'CPI'}, inplace=True)      #  
CPI_data[u''] = [item.timetuple()[0] for item in CPI_data[u'']]           #  

RS_data['DATE'] = pd.to_datetime(RS_data['DATE'], format = '%Y.%m.%d')
RS_data.rename(columns= {'DATE': u'', 'RSAFS': 'TOTAL SALES'}, inplace=True)
RS_data[u''] = [item.timetuple()[0] for item in RS_data[u'']]

fig, ax = plt.subplots(2, 1, figsize = (18, 18), dpi=100)

plt.subplot(2, 1, 1)

plt.plot(CPI_data[u''], CPI_data['CPI'], 'k-o', color='blue', markersize=4, label='CPI')
plt.legend(loc='upper left')
plt.grid()
plt.xlabel(u'')
plt.ylabel('CPI-USA, %')
plt.title(u'     1947  2016 (   )')

plt.subplot(2, 1, 2)
plt.plot(RS_data[u''], RS_data['TOTAL SALES'], 'k-o', color='m', markersize=4, label='TOTAL SALES')
plt.legend(loc='upper left')
plt.grid()
plt.xlabel(u'')
plt.ylabel('RS-USA, mln.dollars')
plt.title(u'      1992  2016 (  )')

plt.show
```
